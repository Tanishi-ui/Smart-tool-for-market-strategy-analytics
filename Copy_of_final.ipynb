```python
def recommend_phone(budget, preference):
    price_threshold = 50000
    iphone_score = 0
    samsung_score = 0

    # Budget logic
    if budget < price_threshold:
        samsung_score += 2
    else:
        iphone_score += 2

    # Normalize input
    preference = preference.lower()

    # Tokenize preferences (support multiple keywords)
    tokens = preference.split()

    # Weight preferences
    for word in tokens:
        if word in ["5g", "technology", "speed", "budget", "performance"]:
            samsung_score += 1
        elif word in ["brand", "discount", "premium", "status", "luxury"]:
            iphone_score += 1

    # Print debug scores (optional)
    print(f"iPhone Score: {iphone_score}, Samsung Score: {samsung_score}")

    # Final decision
    if iphone_score > samsung_score:
        return "Recommended: iPhone (Apple)"
    elif samsung_score > iphone_score:
        return "Recommended: Samsung Galaxy"
    else:
        return "Both are good options based on your inputs."

# User input
budget = float(input("Enter your budget (in INR): "))
preference = input("What do you value most? (You can type multiple: 5G / Technology / Budget / Brand / Discount / Premium): ")

# Get recommendation
result = recommend_phone(budget, preference)
print(result)
```
